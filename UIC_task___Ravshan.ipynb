## Load dataset

```python
from datasets import load_dataset

train_ds = load_dataset("mozilla-foundation/common_voice_17_0", "uz", use_auth_token=True, split='train+validation')
test_ds = load_dataset("mozilla-foundation/common_voice_17_0", "uz", use_auth_token=True, split='test')

def prepare_datasets(batch):
  transcription = batch["sentence"]

  if transcription.startswith('"') and transcription.endswith('"'):
    transcription = transcription[1:-1]

  if len(transcription) > 0 and transcription[-1] not in [".", "?", "!"]:
    transcription = transcription + "."

  batch["sentence"] = transcription

  return batch

train_ds = train_ds.map(prepare_datasets, desc="preprocess dataset")
test_ds = test_ds.map(prepare_datasets)
```

```python
def replace_cyrilic_characters(dataset):
    dataset["sentence"] = re.sub('[а]', 'a', dataset["sentence"])
    dataset["sentence"] = re.sub('[б]', 'b', dataset["sentence"])
    dataset["sentence"] = re.sub('[в]', 'v', dataset["sentence"])
    dataset["sentence"] = re.sub('[г]', 'g', dataset["sentence"])
    dataset["sentence"] = re.sub('[д]', 'd', dataset["sentence"])
    dataset["sentence"] = re.sub('[е]', 'ye', dataset["sentence"])
    dataset["sentence"] = re.sub('[ж]', 'j', dataset["sentence"])
    dataset["sentence"] = re.sub('[и]', 'i', dataset["sentence"])
    dataset["sentence"] = re.sub('[й]', 'y', dataset["sentence"])
    dataset["sentence"] = re.sub('[к]', 'k', dataset["sentence"])
    dataset["sentence"] = re.sub('[л]', 'l', dataset["sentence"])
    dataset["sentence"] = re.sub('[м]', 'm', dataset["sentence"])
    dataset["sentence"] = re.sub('[н]', 'n', dataset["sentence"])
    dataset["sentence"] = re.sub('[о]', 'o', dataset["sentence"])
    dataset["sentence"] = re.sub('[п]', 'p', dataset["sentence"])
    dataset["sentence"] = re.sub('[р]', 'r', dataset["sentence"])
    dataset["sentence"] = re.sub('[с]', 's', dataset["sentence"])
    dataset["sentence"] = re.sub('[т]', 't', dataset["sentence"])
    dataset["sentence"] = re.sub('[у]', 'u', dataset["sentence"])
    dataset["sentence"] = re.sub('[х]', 'x', dataset["sentence"])
    dataset["sentence"] = re.sub('[ч]', 'ch', dataset["sentence"])
    dataset["sentence"] = re.sub('[ш]', 'sh', dataset["sentence"])
    dataset["sentence"] = re.sub('[ъ]', '', dataset["sentence"])
    dataset["sentence"] = re.sub('[ы]', 'i', dataset["sentence"])
    dataset["sentence"] = re.sub('[ь]', '', dataset["sentence"])
    dataset["sentence"] = re.sub('[э]', 'e', dataset["sentence"])
    dataset["sentence"] = re.sub('[ў]', 'oʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[ò]', 'oʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[ó]', 'oʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[ғ]', 'gʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[қ]', 'q', dataset["sentence"])
    dataset["sentence"] = re.sub('[ҳ]', 'h', dataset["sentence"])
    dataset["sentence"] = re.sub('[õ]', 'oʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[ʙ]', 'v', dataset["sentence"])
    dataset["sentence"] = re.sub('[ş]', 'sh', dataset["sentence"])
    dataset["sentence"] = re.sub('[ü]', 'u', dataset["sentence"])
    dataset["sentence"] = re.sub('[´]', 'ʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[’]', 'ʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[`]', 'ʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[’]', 'ʻ', dataset["sentence"])
    dataset["sentence"] = re.sub('[—]', '―', dataset["sentence"])
    return dataset
```

### Prepare the dataset

```python
from transformers import Wav2Vec2CTCTokenizer

tokenizer = Wav2Vec2CTCTokenizer.from_pretrained("./", unk_token="[UNK]", pad_token="[PAD]", word_delimiter_token="|")
```

```python
from transformers import Wav2Vec2FeatureExtractor

feature_extractor = Wav2Vec2FeatureExtractor(feature_size=1, sampling_rate=16000, padding_value=0.0, do_normalize=True, return_attention_mask=True)
```

```python
from transformers import Wav2Vec2Processor

processor = Wav2Vec2Processor(feature_extractor=feature_extractor, tokenizer=tokenizer)
```

```python
def prepare_dataset(batch):
    audio = batch["audio"]

    batch["input_values"] = processor(audio["array"], sampling_rate=audio["sampling_rate"]).input_values[0]
    batch["input_length"] = len(batch["input_values"])

    with processor.as_target_processor():
        batch["labels"] = processor(batch["sentence"]).input_ids
    return batch
```

```python
sample_train_ds = sample_train_ds.map(prepare_dataset, remove_columns=sample_train_ds.column_names)
```

```python
test = test.map(prepare_dataset, remove_columns=test.column_names)
```

```python
import torch

from dataclasses import dataclass, field
from typing import Any, Dict, List, Optional, Union

@dataclass
class DataCollatorCTCWithPadding:

    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:

        input_features = [{"input_values": feature["input_values"]} for feature in features]
        label_features = [{"input_ids": feature["labels"]} for feature in features]

        batch = self.processor.pad(
            input_features,
            padding=self.padding,
            return_tensors="pt",
        )
        with self.processor.as_target_processor():
            labels_batch = self.processor.pad(
                label_features,
                padding=self.padding,
                return_tensors="pt",
            )

        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)

        batch["labels"] = labels

        return batch
```

```python
data_collator = DataCollatorCTCWithPadding(processor=processor, padding=True)
```

## Training the model

```python
from datasets import load_metric

wer_metric = load_metric("wer")
```

```python
def compute_metrics(pred):
    pred_logits = pred.predictions
    pred_ids = np.argmax(pred_logits, axis=-1)

    pred.label_ids[pred.label_ids == -100] = processor.tokenizer.pad_token_id

    pred_str = processor.batch_decode(pred_ids)
    label_str = processor.batch_decode(pred.label_ids, group_tokens=False)

    wer = wer_metric.compute(predictions=pred_str, references=label_str)

    return {"wer": wer}
```

```python
from transformers import Wav2Vec2ForCTC

model = Wav2Vec2ForCTC.from_pretrained(
    "facebook/wav2vec2-xls-r-300m",
    attention_dropout=0.0,
    hidden_dropout=0.0,
    feat_proj_dropout=0.0,
    mask_time_prob=0.05,
    layerdrop=0.0,
    ctc_loss_reduction="mean",
    pad_token_id=processor.tokenizer.pad_token_id,
    vocab_size=len(processor.tokenizer),
)
```

```python
model.freeze_feature_extractor()
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
  output_dir="/content/drive/MyDrive/pytorch_model.bin",
  group_by_length=True,
  per_device_train_batch_size=16,
  gradient_accumulation_steps=2,
  evaluation_strategy="steps",
  num_train_epochs=15,
  gradient_checkpointing=True,
  fp16=True,
  save_steps=400,
  eval_steps=400,
  logging_steps=400,
  learning_rate=3e-4,
  warmup_steps=500,
  save_total_limit=2,
  push_to_hub=True,
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    data_collator=data_collator,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=sample_train_ds,
    eval_dataset=test,
    tokenizer=processor.feature_extractor,
)
```

```python
trainer.train()
```
